```python
import rqdatac
rqdatac.init()
```

## 沪深300

```python
from rqdatac import get_price

hs300_df = get_price(order_book_ids='000300.XSHG', start_date='2012-01-01', end_date='2022-03-22', fields='close')
hs300_df.reset_index().pivot(index='date', columns='order_book_id', values='close').plot(grid=True)
```
